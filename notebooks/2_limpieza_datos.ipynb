### **Limpieza de datos**

```python
# Cargar las librerias necesarias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset 
df_tortilla_prices = pd.read_csv('../data/raw/tortilla_prices.csv')
df_tortilla_prices.head(15)
```

##### **Eliminar valores nulos**

```python
# Verificar la cantidad de fidas y columnas 
num_fias, num_columnas = df_tortilla_prices.shape
print(f'Número de filas: {num_fias}\nNúmero de columnas: {num_columnas}')
```

```python
# Mostrar las filas con los valores nulos
filas_con_nulos = df_tortilla_prices[df_tortilla_prices.isnull().any(axis=1)]
filas_con_nulos
```

```python
# Eliminar las filas con los valores nulos
df_tortilla_prices_sin_nulos = df_tortilla_prices.dropna()

# Verificar la cantidad de filas y columnas
num_filas_sin_nulos, num_columnas_sin_nulos = df_tortilla_prices_sin_nulos.shape
print(f'Número de filas: {num_filas_sin_nulos}\nNúmero de columnas: {num_columnas_sin_nulos}')
```

```python
# Verificar si quedan valores nulos
num_nulos = df_tortilla_prices_sin_nulos.isnull().sum()
print(f'Números de valores nulos:\n{num_nulos}')
```

Antes de eliminar los valores nulos el DataFrame contenia 289146 orginalmente.

Despues de eliminar los valores nulos el DataFrame contiene 282756.

Se eliminaron 6390 filas con valores nulos.
